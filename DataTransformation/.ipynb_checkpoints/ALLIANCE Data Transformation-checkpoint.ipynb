```python
import pandas as pd
import numpy as np
```

```python
!ls ../SourceData/CorrelatesOfWar/
```

```python
dfAllianceAll = pd.read_csv('../SourceData/CorrelatesOfWar/alliance_v4.1_by_member.csv')
dfAllianceAll.columns
```

## Create 'STATE_ALLIANCE' table

task: transform 'alliance_v4.1_by_member.csv' (saved as 'dfAllianceAll') into a table with the following attributes:

- AllianceID
- AllianceType
- StartDate
- EndDate
- StartYear
- StartMonth
- StartDay
- EndYear
- EndMonth
- EndDay

```python
dfAlliance = dfAllianceAll[['version4id', 'all_st_day', 'all_st_month', 'all_st_year', 'all_end_day', 'all_end_month', 'all_end_year', 'ss_type']]
dfAlliance.rename(columns={'version4id':'AllianceID', 'all_st_day':'StartDay', 'all_st_month':'StartMonth',
       'all_st_year':'StartYear', 'all_end_day':'EndDay', 'all_end_month':'EndMonth', 'all_end_year':'EndYear',
       'ss_type':'AllianceType'}, inplace=True)
dfAlliance.drop_duplicates(inplace=True)
dfAlliance = dfAlliance.fillna('')
dfAlliance
```

```python
dfAlliance['StartDate'] = pd.to_datetime(dict(year=dfAlliance.StartYear, month=dfAlliance.StartMonth, day=dfAlliance.StartDay), format='%Y%m%d', errors='ignore')
dfAlliance['EndDate'] = pd.to_datetime(dict(year=dfAlliance.EndYear, month=dfAlliance.EndMonth, day=dfAlliance.EndDay), errors='coerce')
dfAlliance['StartDate'] = dfAlliance['StartDate'].apply(lambda x: x.strftime('%Y-%m-%d'))
dfAlliance['EndDate'] = dfAlliance['EndDate'].apply(lambda x: x.strftime('%Y-%m-%d') if pd.notnull(x) else '')
dfAlliance
```

```python
dfAlliance = dfAlliance[['AllianceID', 'AllianceType', 'StartDate', 'EndDate', 'StartYear', 'StartMonth', 'StartDay', 'EndYear', 'EndMonth', 'EndDay']]
dfAlliance
```

```python
dfAlliance.to_csv('../FinalData/state_alliance.csv', encoding='utf-8', index=False)
```

## Create 'ALLIANCE_MEMBERSHIP' table

task: transform 'alliance_v4.1_by_member.csv' (saved as 'dfAllianceAll') into a table with the following attributes:

- AllianceID
- StateID
- StartDate
- EndDate
- StartYear
- StartMonth
- StartDay
- EndYear
- EndMonth
- EndDay

```python
dfAllianceMem = dfAllianceAll[['version4id', 'ccode', 'mem_st_day', 'mem_st_month', 'mem_st_year', 'mem_end_day', 'mem_end_month', 'mem_end_year', 'defense', 'neutrality', 'nonaggression', 'entente']]
dfAllianceMem.rename(columns={'version4id':'AllianceID', 'ccode':'StateID', 'mem_st_day':'StartDay', 
                              'mem_st_month':'StartMonth', 'mem_st_year':'StartYear', 'mem_end_day':'EndDay', 
                              'mem_end_month':'EndMonth', 'mem_end_year':'EndYear'}, inplace=True)
dfAllianceMem.drop_duplicates(inplace=True)
dfAllianceMem = dfAllianceMem.fillna('')
dfAllianceMem
```

```python
dfAllianceMem['StartDate'] = pd.to_datetime(dict(year=dfAllianceMem.StartYear, month=dfAllianceMem.StartMonth, day=dfAllianceMem.StartDay), format='%Y%m%d', errors='ignore')
dfAllianceMem['EndDate'] = pd.to_datetime(dict(year=dfAllianceMem.EndYear, month=dfAllianceMem.EndMonth, day=dfAllianceMem.EndDay), errors='coerce')
dfAllianceMem['StartDate'] = dfAllianceMem['StartDate'].apply(lambda x: x.strftime('%Y-%m-%d'))
dfAllianceMem['EndDate'] = dfAllianceMem['EndDate'].apply(lambda x: x.strftime('%Y-%m-%d') if pd.notnull(x) else '')
dfAllianceMem
```

```python
dfAllianceMembership = dfAllianceMem[['AllianceID', 'StateID', 'StartDate', 'EndDate', 'StartYear', 'StartMonth', 'StartDay', 'EndYear', 'EndMonth', 'EndDay']]
dfAllianceMembership
```

```python
dfAllianceMembership.to_csv('../FinalData/alliance_membership.csv', encoding='utf-8', index=False)
```

## Create 'ALLIANCE_TRAITS' table

task: transform 'alliance_v4.1_by_member.csv' (saved as 'dfAllianceAll') into a table with the following attributes:

- AllianceID
- StateID
- StartDate
- Trait

```python
dfAllianceTraits = dfAllianceMem[['AllianceID', 'StateID', 'StartDate', 'defense', 'neutrality', 'nonaggression', 'entente']]
dfAllianceTraits.replace('', 0, inplace=True)
dfAllianceTraits
```

```python
dfAllianceTraits['defense'] [dfAllianceTraits['defense'] == 1] = 'defense'
dfAllianceTraits['neutrality'] [dfAllianceTraits['neutrality'] == 1] = 'neutrality'
dfAllianceTraits['nonaggression'] [dfAllianceTraits['nonaggression'] == 1] = 'nonaggression'
dfAllianceTraits['entente'] [dfAllianceTraits['entente'] == 1] = 'entente'
dfAllianceTraits['defense'] [dfAllianceTraits['defense'] == 0 ] = None
dfAllianceTraits['neutrality'] [dfAllianceTraits['neutrality'] == 0] = None
dfAllianceTraits['nonaggression'] [dfAllianceTraits['nonaggression'] == 0] = None
dfAllianceTraits['entente'] [dfAllianceTraits['entente'] == 0] = None
dfAllianceTraits
```

```python
dfAllianceTraitsD = dfAllianceTraits[['AllianceID', 'StateID', 'StartDate', 'defense']]
dfAllianceTraitsN = dfAllianceTraits[['AllianceID', 'StateID', 'StartDate', 'neutrality']]
dfAllianceTraitsA = dfAllianceTraits[['AllianceID', 'StateID', 'StartDate', 'nonaggression']]
dfAllianceTraitsE = dfAllianceTraits[['AllianceID', 'StateID', 'StartDate', 'entente']]

dfAllianceTraitsD = dfAllianceTraitsD.dropna(subset=['defense'])
dfAllianceTraitsN = dfAllianceTraitsN.dropna(subset=['neutrality'])
dfAllianceTraitsA = dfAllianceTraitsA.dropna(subset=['nonaggression'])
dfAllianceTraitsE = dfAllianceTraitsE.dropna(subset=['entente'])

dfAllianceTraitsD.rename(columns={'defense':'Trait'}, inplace=True)
dfAllianceTraitsN.rename(columns={'neutrality':'Trait'}, inplace=True)
dfAllianceTraitsA.rename(columns={'nonaggression':'Trait'}, inplace=True)
dfAllianceTraitsE.rename(columns={'entente':'Trait'}, inplace=True)
```

```python
alltraits = [dfAllianceTraitsD, dfAllianceTraitsN, dfAllianceTraitsA, dfAllianceTraitsE]

dfAllianceAllTraits = pd.concat(alltraits)
dfAllianceAllTraits
```

```python
dfAllianceAllTraits.to_csv('../FinalData/alliance_traits.csv', encoding='utf-8', index=False)
```
